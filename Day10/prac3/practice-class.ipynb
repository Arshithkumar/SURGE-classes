```python
import pandas as pd
import numpy as np
```

```python
meal=pd.read_csv('meal_info.csv')
meal
```

```python
food=pd.read_csv('food.csv')
food
```

```python
fullfill=pd.read_csv('fulfilment_center_info.csv')
fullfill
```

```python
merg1 = pd.merge(food,fullfill,on='center_id',how='inner')
merg1
```

```python
merg2=pd.merge(merg1,meal,on='meal_id',how='inner')
merg2.head()
```

```python
meal.category.value_counts()
```

```python
merg2.groupby('category')[['num_orders']].sum()
```

```python
m=merg2.groupby('category')['num_orders'].sum()/meal['category'].value_counts()
m
```

```python
merg2.shape
```

```python
merg2.index
```

```python
import matplotlib.pyplot as plt
```

```python
table=merg2.pivot_table(values='num_orders',index='category',aggfunc=sum)
```

```python
plt.bar(table.index, table['num_orders'])
plt.xticks(rotation=70)
plt.xlabel('Food item')
plt.ylabel('Quantity of Food')
plt.title('Food Graph')
```
