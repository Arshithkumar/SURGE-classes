```python
list1=[3, 6, 9, 12, 15, 18, 21]
list2=[4, 8, 12, 16, 20, 24, 28]
list3=list()
odd=list1[1::2]
print("Element at odd-index positions from list one")
print(odd)
even=list2[0::2]
print("Element at even-index positions from list two")
print(even)
list3.extend(odd)    #to iterate over an iterable(list) 
list3.extend(even)   #then add each element of the iterable to the end of the current list.
print("Printing Final third list")
print(list3)         #using list slicing
```

```python
sample_list = [11, 45, 8, 23, 14, 12, 78, 45, 89]
print("Original list ", sample_list)
length = len(sample_list)
chunk_size = int(length / 3)
start = 0
end = chunk_size
for i in range(3):
    indexes = slice(start, end)
    list_chunk = sample_list[indexes]
    print("Chunk ", i, list_chunk)
    print("After reversing it ", list(reversed(list_chunk)))
    start = end
    end += chunk_size
```

```python
list1=[11, 45, 8, 11, 23, 45, 23, 45, 89]
print("Original list ",list1)
count=dict()
for item in list1:
    if item in count:
        count[item]+= 1
    else:
        count[item]=1
print("Printing count of each item  ",count)
```

```python
set1={23, 42, 65, 57, 78, 83, 29}
set2={57, 83, 29, 67, 73, 43, 48}
print("First Set ",set1)
print("Second Set ",set2)
intersection=set1.intersection(set2)
print("Intersection is ", intersection)
for item in intersection:
    set1.remove(item)
print("First Set after removing common element ",set1)
```

```python
set1={57, 83, 29}
set2={57, 83, 29, 67, 73, 43, 48}
print("First Set ",set1)
print("Second Set ",set2)
print("First set is subset of second set -",set1.issubset(set2))
print("Second set is subset of First set - ",set2.issubset(set1))
print("First set is Super set of second set - ",set1.issuperset(set2))
print("Second set is Super set of First set - ",set2.issuperset(set1))
if set1.issubset(set2):
    set1.clear()
if set2.issubset(set1):
    set2.clear()
print("First Set ",set1)
print("Second Set ",set2)
```

```python
rno=[47, 64, 69, 37, 76, 83, 95, 97]
sampleDict={'Jhon': 47, 'Emma': 69, 'Kelly': 76, 'Jason': 97}
print("List:",rno)
print("Dictionary:",sampleDict)
rno[:] = [item for item in rno if item in sampleDict.values()]
print("after removing unwanted elements from list:", rno)
```

```python
speed={'jan':47, 'feb':52, 'march':47, 'April':44, 'May':52, 'June':53, 'july':54, 'Aug':44, 'Sept':54}
speed_values=list(speed.values())
unique_sv = []
for i in speed_values:
    if i not in unique_sv:
        unique_sv.append(i)
print(unique_sv)
```
