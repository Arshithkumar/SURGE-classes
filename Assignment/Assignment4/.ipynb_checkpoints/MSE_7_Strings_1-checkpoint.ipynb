```python
#1
def get_middle_three_chars(s):
    mid = len(s) // 2
    return s[mid-1:mid+2]
str1 = "JhonDipPeta"
print(get_middle_three_chars(str1)) 
str2 = "JaSonAy"
print(get_middle_three_chars(str2))  
```

```python
#2
def append_to_middle(s1, s2):
    middle_index = len(s1) // 2
    return s1[:middle_index] + s2 + s1[middle_index:]
s1 = "Ault"
s2 = "Kelly"
print(append_to_middle(s1, s2))  
```

```python
#3
def get_first_middle_last_chars(s1, s2):
    s1_middle_index = len(s1) // 2
    s2_middle_index = len(s2) // 2
    return s1[0] + s2[0] + s1[s1_middle_index] + s2[s2_middle_index] + s1[-1] + s2[-1]
s1 = "America"
s2 = "Japan"
print(get_first_middle_last_chars(s1, s2)) 
```

```python
#4
def arrange_lowercase_first(s):
    lower_chars = ""
    upper_chars = ""
    for char in s:
        if char.islower():
            lower_chars += char
        else:
            upper_chars += char
    return lower_chars + upper_chars
str1 = "PyNaTive"
print(arrange_lowercase_first(str1))  
```

```python
#5
def count_chars_digits_symbols(s):
    char_count = 0
    digit_count = 0
    symbol_count = 0
    for char in s:
        if char.isalpha():
            char_count += 1
        elif char.isdigit():
            digit_count += 1
        else:
            symbol_count += 1
    return (char_count, digit_count, symbol_count)
str1 = "P@#yn26at^&i5ve"
char_count, digit_count, symbol_count = count_chars_digits_symbols(str1)
print("Total counts of chars, digits, and symbols")
print("Chars =", char_count)
print("Digits =", digit_count)
print("Symbol =", symbol_count)
```

```python
#6
def mix_strings(s1, s2):
    result = ""
    i = 0
    j = len(s2) - 1
    while i < len(s1) and j >= 0:
        result += s1[i]
        result += s2[j]
        i += 1
        j -= 1
    if i < len(s1):
        result += s1[i:]
    if j >= 0:
        result += s2[:j+1][::-1]
    return result
s1 = "Abc"
s2 = "Xyz"
print(mix_strings(s1, s2)) 
```

```python
#7
def is_balanced(s1, s2):
    for c in s1:
        if c not in s2:
            return False
    return True
# Case 1:
s1 = "Yn"
s2 = "PYnative"
print(is_balanced(s1, s2))  

# Case 2:
s1 = "Ynf"
s2 = "PYnative"
print(is_balanced(s1, s2)) 
```

```python
#8
def count_usa(str1):
    return str1.lower().count('usa')
str1 = "Welcome to USA. usa awesome, isn't it?"
print("The USA count is:", count_usa(str1)) 
```

```python
#9
import re

def sum_and_avg_digits(str1):
    digits = [int(d) for d in re.findall('\d+', str1)]
    sum_digits = sum(digits)
    avg_digits = sum_digits / len(digits)
    return sum_digits, avg_digits
str1 = "English = 78 Science = 83 Math = 68 History = 65"
sum_digits, avg_digits = sum_and_avg_digits(str1)
print("sum is", sum_digits)
print("average is", avg_digits)
```

```python
#10
def count_chars(str1):
    char_counts = {}
    for char in str1:
        if char in char_counts:
            char_counts[char] += 1
        else:
            char_counts[char] = 1
    return char_counts
str1 = "Apple"
char_counts = count_chars(str1)
print(char_counts)
```

```python
#11
def reverse_string(str1):
    return str1[::-1]
str1 = "PYnative"
reversed_str = reverse_string(str1)
print(reversed_str)
```

```python
#12
def find_last_position(str1, sub):
    last_pos = str1.rfind(sub)
    return last_pos

str1 = "Emma is a data scientist who knows Python. Emma works at google."
sub = "Emma"
last_pos = find_last_position(str1, sub)
print("Last occurrence of", sub, "starts at index", last_pos)
```

```python
#13
def split_string(str1):
    substrings = str1.split("-")
    for substring in substrings:
        print(substring)

str1 = "Emma-is-a-data-scientist"
print("Displaying each substring")
split_string(str1)
```

```python
#14
def remove_empty_strings(str_list):
    print("Original list of strings")
    print(str_list)
    str_list = [string for string in str_list if string and string.strip()]

    print("After removing empty strings")
    print(str_list)

str_list = ["Emma", "Jon", "", "Kelly", None, "Eric", ""]
remove_empty_strings(str_list)
```

```python
#18
import string
str1 = '/*Jon is @developer & musician!!'
punctuations = string.punctuation
new_str = ""
for char in str1:
    if char in punctuations:
        new_str += "#"
    else:
        new_str += char
print(new_str)
```
