```python
#01
from IPython.display import display
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

# importing required packages
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
```

```python
# extracting the required dataset 
df = pd.read_csv('Stock_broker_Quarter_Analysis.csv')
df
```

```python
# returning the type, shape, column names and some sample data from the extracted dataset

print("`type` of `df` :",end = "\n")
type(df)


print("\n* * * * * * * *\n")


print("`shape` of `df` :",end = "\n")
df.shape


print("\n* * * * * * * *\n")


print("listing the column names of the `df` :",end = "\n")
df.columns.to_list()


print("\n* * * * * * * *\n")


print("Returning sample `df` :",end = "\n")
df.head(3)
```

```python
# assigning and droping 'unique_variable(s)' and 'dependent-variable' from the dataframe

unique_variable = df['Stock_broker_id']
target_variable = df['Succesful']

df.drop(['Stock_broker_id','Succesful'],axis=1,inplace=True)
```

```python
# preprocessing `df`...

## splitting `df` based on the 'dtypes':
df_numerical = df.select_dtypes(include = ['number'])
df_categorical = df.select_dtypes(include = ['object'])

## imputing the missing data at `df_numerical`:
imputer_mdn = SimpleImputer(strategy = 'median')
df_num_mdn_imputer = pd.DataFrame(imputer_mdn.fit_transform(df_numerical))
df_num_mdn_imputer.columns = df_numerical.columns
df_num_mdn_imputer.index = df_numerical.index
df_numerical = df_num_mdn_imputer

## pruning garbage variables from `df_categorical`:
df_categorical.drop(['Name','Seating_Location'],axis=1,inplace=True)

## encoding the categorical data at `df_categorical`:
lbl_enc = LabelEncoder()
df_categorical[df_categorical.columns.to_list()] = df_categorical[df_categorical.columns.to_list()].apply(lbl_enc.fit_transform)

## merging the `df_numerical` & `df_catagerical` into `df`:
df = pd.concat([df_numerical,df_categorical],axis=1)
#df
```

```python
# splitting into train-test
x_train, x_test, y_train, y_test = train_test_split(df, target_variable, test_size=0.25)
```

```python
def func_model_dtc(param__x_train, param__x_test, param__y_train, param__y_test,param__max_depth): 
    model_dtc = DecisionTreeClassifier(max_depth = param__max_depth)
    dtc_rfe = RFE(model_dtc)
    dtc_rfe.fit(param__x_train,param__y_train)
    dtc_rfe_labels = param__x_train.columns.to_list()
    dtc_rfe_ranks = dtc_rfe.ranking_
    df_dtc_rfe_features = pd.DataFrame(sorted(list(map(lambda xt, yt: (xt, yt), dtc_rfe_ranks, dtc_rfe_labels))),columns=['rank','features'])
    dtc_rfe_selected_features = df_dtc_rfe_features['features'][df_dtc_rfe_features['rank'] == 1]
    dtc_rfe_selected_features
    x_train, x_test = param__x_train[dtc_rfe_selected_features], param__x_test[dtc_rfe_selected_features]
    model_dtc.fit(x_train,y_train)
    pred__y_test = model_dtc.predict(x_test)
    pred__y_train = model_dtc.predict(x_train)
    accy_scr__train =  accuracy_score(y_train,pred__y_train)
    accy_scr__test = accuracy_score(y_test,pred__y_test)
    return(param__max_depth,accy_scr__train,accy_scr__test)
lst__init_max_depth = [4,6,8,10]
lst__max_depth = []
lst__accy_src_train = []
lst__accy_src_test = []

for itr__max_depth in lst__init_max_depth:
    rtrn__func_model_dtc = func_model_dtc(x_train, x_test, y_train, y_test,itr__max_depth)
    lst__max_depth.append(rtrn__func_model_dtc[0])
    lst__accy_src_train.append(rtrn__func_model_dtc[1]*100)
    lst__accy_src_test.append(rtrn__func_model_dtc[2]*100)
```

```python
# Using/Initiating 'max_depth' values as '[4,6,8,10]':
max_depth = lst__max_depth
print(f"Initiated 'max-depth' values : {max_depth}")


print("\n* * * * * * * *\n")


# Storing training accuracies in `training_results` ...
training_results = lst__accy_src_train
print(f"Accuracies of training values : {training_results}")


print("\n* * * * * * * *\n")


# Storing test accuracies in `test_results` ...
test_results = lst__accy_src_test
print(f"Accuracies of testing values : {test_results}")


print("\n* * * * * * * *\n")



df__model_dtc = pd.DataFrame({'max_depth' : max_depth, 'training_results' : training_results, 'test_results' : test_results})

#  Plotting `training_results` w.r.t `max_depth` ...
print("Plotting `training_results` w.r.t `max_depth` ...",end="\n")
plt.plot(df__model_dtc['max_depth'],df__model_dtc['training_results'])
plt.title("'max_depth' vs 'training_results'")
plt.xlabel('max_depth')
plt.ylabel('training_results')
plt.show()


print("\n* * * * * * * *\n")


# Plotting `test_results` w.r.t `max_depth` ...
print("Plotting `test_results` w.r.t `max_depth` ...",end="\n")
plt.plot(df__model_dtc['max_depth'],df__model_dtc['test_results'])
plt.title("'max_depth' vs 'test_results'")
plt.xlabel('max_depth')
plt.ylabel('test_results')
plt.show()
```

```python
#02
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import RFE
from matplotlib import pyplot as plt
from sklearn.metrics import confusion_matrix
df = pd.read_csv('Stock_broker_Quarter_Analysis.csv')
df
```

```python
# returning the type, shape, column names and some sample data from the extracted dataset ...

print("`type` of `df` :",end = "\n")
type(df)


print("\n* * * * * * * *\n")


print("`shape` of `df` :",end = "\n")
df.shape


print("\n* * * * * * * *\n")


print("listing the column names of the `df` :",end = "\n")
df.columns.to_list()


print("\n* * * * * * * *\n")


print("Returning sample `df` :",end = "\n")
df.head(3)
```

```python
# assigning and droping 'unique_variable(s)' and 'dependent-variable' from the `df`:

unique_variable = df['Stock_broker_id']
target_variable = df['Succesful']

df.drop(['Stock_broker_id','Succesful'],axis=1,inplace=True)
```

```python
# preprocessing `df`...

df_numerical = df.select_dtypes(include = ['number'])
df_categorical = df.select_dtypes(include = ['object'])
#imputation
imputer_mdn = SimpleImputer(strategy = 'median')
df_num_mdn_imputer = pd.DataFrame(imputer_mdn.fit_transform(df_numerical))
df_num_mdn_imputer.columns = df_numerical.columns
df_num_mdn_imputer.index = df_numerical.index
df_numerical = df_num_mdn_imputer
## pruning 
df_categorical.drop(['Name','Seating_Location'],axis=1,inplace=True)
## encoding categorical data 
lbl_enc = LabelEncoder()
df_categorical[df_categorical.columns.to_list()] = df_categorical[df_categorical.columns.to_list()].apply(lbl_enc.fit_transform)
## merging the `df_numerical` & `df_catagerical` into `df`:
df = pd.concat([df_numerical,df_categorical],axis=1)
#df
```

```python
# splitting into train-test
x_train, x_test, y_train, y_test = train_test_split(df, target_variable, test_size=0.25)
```

```python
# initializing 'DecisionTreeClassifier' w.r.t the given 'max-depth = 8' based on the feature-selection and storing the 'train & test' accuracy of the model
model_dtc = DecisionTreeClassifier(max_depth = 8)
dtc_rfe = RFE(model_dtc)
dtc_rfe.fit(x_train,y_train)
dtc_rfe_labels = x_train.columns.to_list()
dtc_rfe_ranks = dtc_rfe.ranking_
df_dtc_rfe_features = pd.DataFrame(sorted(list(map(lambda xt, yt: (xt, yt), dtc_rfe_ranks, dtc_rfe_labels))),columns=['rank','features'])
dtc_rfe_selected_features = df_dtc_rfe_features['features'][df_dtc_rfe_features['rank'] == 1]
x_train, x_test = x_train[dtc_rfe_selected_features], x_test[dtc_rfe_selected_features]
model_dtc.fit(x_train,y_train)
pred__y_train = model_dtc.predict(x_train)
pred__y_test = model_dtc.predict(x_test)
```

```python
# initalizing the scaling scores of `feature_inportances` of the 'selected features' ...
fte_imp = model_dtc.feature_importances_

# Accessing feature importances from the model created ...
print(f"Accessing feature importances from the model created :",end= "\n")
display(fte_imp)


print("\n* * * * * * * *\n")



#. Plotting the 'Bar Chart' reflecting importances of each feature with proper label ...
plt.bar(dtc_rfe_selected_features, fte_imp)
plt.title('RFE Selection Feature Importances')
plt.xlabel('Selection Features')
plt.ylabel('Selection Importance')
plt.show()


print("\n* * * * * * * *\n")


# Calculating the `confusion_matrix` for the test data ...
test_TN, test_FP, test_FN, test_TP = confusion_matrix(y_test,pred__y_test).ravel()

# Computing accuracy, recall, specificity values for the model created and storing in a dictionary.(test_data)
test_accuracy = (test_TN + test_TP) / (test_TN + test_FP + test_FN + test_TP)
test_recall = (test_TP) / (test_TP + test_FN)
test_specificity = (test_TN) / (test_TN + test_FP)
test_dict__acry_rcl_specty = {
                                'test_accuracy' : test_accuracy,
                                'test_recall' : test_recall,
                                'test_specificity' : test_specificity
                                
                              }
print("Returning the stored 'accuracy, recall, specificity' values of test data :",end = "\n")
test_dict__acry_rcl_specty



print("\n* * * * * * * *\n")


# Calculating the `confusion_matrix` for the train data ...
train_TN, train_FP, train_FN, train_TP = confusion_matrix(y_train,pred__y_train).ravel()

# Computing accuracy, recall, specificity values for the model created and storing in a dictionary.(training_data)
train_accuracy = (train_TN + train_TP) / (train_TN + train_FP + train_FN + train_TP)
train_recall = (train_TP) / (train_TP + train_FN)
train_specificity = (train_TN) / (train_TN + train_FP)
train_dict__acry_rcl_specty = {
                                'train_accuracy' : train_accuracy,
                                'train_recall' : train_recall,
                                'train_specificity' : train_specificity
                                
                              }
print("Returning the stored 'accuracy, recall, specificity' values of train data :",end = "\n")
train_dict__acry_rcl_specty
```

```python
# assigning and droping 'unique_variable(s)' and 'dependent-variable' from the `df`:

unique_variable = df['Stock_broker_id']
target_variable = df['Succesful']

df.drop(['Stock_broker_id','Succesful'],axis=1,inplace=True)
```

```python
# preprocessing `df`...

## splitting `df` based on the 'dtypes':
df_numerical = df.select_dtypes(include = ['number'])
df_categorical = df.select_dtypes(include = ['object'])

## imputing the missing data at `df_numerical`:
imputer_mdn = SimpleImputer(strategy = 'median')
df_num_mdn_imputer = pd.DataFrame(imputer_mdn.fit_transform(df_numerical))
df_num_mdn_imputer.columns = df_numerical.columns
df_num_mdn_imputer.index = df_numerical.index
df_numerical = df_num_mdn_imputer

## pruning garbage variables from `df_categorical`:
df_categorical.drop(['Name','Seating_Location'],axis=1,inplace=True)

## encoding the categorical data at `df_categorical`:
lbl_enc = LabelEncoder()
df_categorical[df_categorical.columns.to_list()] = df_categorical[df_categorical.columns.to_list()].apply(lbl_enc.fit_transform)

## merging the `df_numerical` & `df_catagerical` into `df`:
df = pd.concat([df_numerical,df_categorical],axis=1)
#df
```

```python
# splitting into train-test
x_train, x_test, y_train, y_test = train_test_split(df, target_variable, test_size=0.25)
```

```python
# initializing 'DecisionTreeClassifier' w.r.t the given 'max-depth = 8' based on the feature-selection and storing the 'train & test' accuracy of the model...
model_dtc = DecisionTreeClassifier(max_depth = 8)
dtc_rfe = RFE(model_dtc)
dtc_rfe.fit(x_train,y_train)
dtc_rfe_labels = x_train.columns.to_list()
dtc_rfe_ranks = dtc_rfe.ranking_
df_dtc_rfe_features = pd.DataFrame(sorted(list(map(lambda xt, yt: (xt, yt), dtc_rfe_ranks, dtc_rfe_labels))),columns=['rank','features'])
dtc_rfe_selected_features = df_dtc_rfe_features['features'][df_dtc_rfe_features['rank'] == 1]


print("\n* * * * * * * *\n")


# Accessing feature importances from the model created ...
print("Accessing feature importances from the model created :",end="\n")
dtc_rfe_selected_features


print("\n* * * * * * * *\n")


# initializing the selected features to the `x_train & x_test` variables ...
x_train, x_test = x_train[dtc_rfe_selected_features], x_test[dtc_rfe_selected_features]

# 'fitting' the model with the 'train' data ...
model_dtc.fit(x_train,y_train)
```

```python
# Accessing predicted probabilities from the model created for each observation of test data ...
true_prob__pred_y_test = model_dtc.predict_proba(x_test)[:,1]
print("Accessing predicted probabilities from the model created for each observation of test data ('positive class'): ",end = "\n")
true_prob__pred_y_test


print("\n* * * * * * * *\n")



# Creating a dataframe with true value, predicted probability of the test data ...
df = pd.DataFrame({'true_val' : y_test,'true_prob__pred_y_test' : true_prob__pred_y_test})
print("Returning the `df` with true value, predicted probability of the test data :")
df



# Adding a variable â€˜adjusted_prediction_02â€™ to that dataframe with values as 1 if predicted probability greater than or equal to 0.2 and as 0 if less than 0.2 (cut_off=0.2) ...
df['adjusted_prediction_02'] = [1 if(itr >= 0.2) else 0 if(itr < 0.2) else None for itr in df['true_prob__pred_y_test']]

# Adding a variable â€˜adjusted_prediction_015â€™ to that dataframe with values as 1 if predicted probability greater than or equal to 0.15 and as 0 if less than 0.15 (cut_off=0.15) ...
df['adjusted_prediction_015'] = [1 if(itr >= 0.15) else 0 if(itr < 0.15) else None for itr in df['true_prob__pred_y_test']] 

# Adding a variable â€˜adjusted_prediction_025â€™ to that dataframe with values as 1 if predicted probability greater than or equal to 0.25 and as 0 if less than 0.25 (cut_off=0.25) ...
df['adjusted_prediction_025'] = [1 if(itr >= 0.25) else 0 if(itr < 0.25) else None for itr in df['true_prob__pred_y_test']]

#  Adding a variable â€˜adjusted_prediction_05â€™ to that dataframe with values as 1 if predicted probability greater than or equal to 0.5 and as 0 if less than 0.5 (cut_off=0.5) ...
df['adjusted_prediction_05'] = [1 if(itr >= 0.5) else 0 if(itr < 0.5) else None for itr in df['true_prob__pred_y_test']]

print("\n* * * * * * * *\n")



# Compute accuracy, recall, specificity using true values and  ['adjusted_prediction_02','adjusted_prediction_015','adjusted_prediction_025','adjusted_prediction_05'] ...
def func__val_prob_score(y_test,df__col_prob):
    val_prob__TN, val_prob__FP, val_prob__FN, val_prob__TP = confusion_matrix(y_test,df[df__col_prob]).ravel()

    val_prob__accuracy = (test_TN + val_prob__TP) / (val_prob__TN + val_prob__FP + val_prob__FN + val_prob__TP)
    val_prob__recall = (test_TP) / (test_TP + test_FN)
    val_prob__specificity = (test_TN) / (test_TN + test_FP)
    
    return (val_prob__accuracy,val_prob__recall,val_prob__specificity)

lst__cut_off = [0.20,0.15,0.25,0.50]
lst__df_col_prob = ['adjusted_prediction_02','adjusted_prediction_015','adjusted_prediction_025','adjusted_prediction_05']
lst__val_prob__accuracy = []
lst__val_prob__recall = []
lst__val_prob__specificity = []

for itr_prob_col in lst__df_col_prob:
    rtrn_func__val_prob_score = func__val_prob_score(y_test,itr_prob_col)
    lst__val_prob__accuracy.append(rtrn_func__val_prob_score[0])
    lst__val_prob__recall.append(rtrn_func__val_prob_score[1])
    lst__val_prob__specificity.append(rtrn_func__val_prob_score[2])




# Creating a dataframe with cut_off values, accuracy values, recall values, specificity values in each columns from above computations ...
val_prob__dict__acry_rcl_specty = {
                                'val_prob__cutoff' : lst__cut_off,
                                'df_col__prob' : lst__df_col_prob,
                                'val_prob__accuracy' : lst__val_prob__accuracy,
                                'val_prob__recall' : lst__val_prob__recall,
                                'val_prob__specificity' : lst__val_prob__specificity
                                
                              }
df__val_prob = pd.DataFrame(val_prob__dict__acry_rcl_specty)
print("Returning `df__val_prob` :")
df__val_prob
```
