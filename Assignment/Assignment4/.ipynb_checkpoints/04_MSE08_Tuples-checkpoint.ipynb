```python
tuple1=(11,22)
tuple2=(99,88)
tuple1,tuple2=tuple2,tuple1
print("tuple1 =",tuple1)
print("tuple2 =",tuple2)
```

```python
tuple1=(11,22,33,44,55,66)
tuple2=tuple1[3:5]
print(tuple2)
```

```python
tuple1=(11, [22, 33], 44, 55)
tuple1[1][0]=222
print(tuple1)
```

```python
tuple1=(('a', 23),('b', 37),('c', 11), ('d',29))
tuple1=tuple(sorted(list(tuple1),key=lambda x: x[1]))
print(tuple1)
```

```python
tuple1=(50, 10, 60, 70, 50)
c=tuple1.count(50)
print(c)
```

```python
def check(t):
    return all(i == t[0] for i in t)
tuple1=(45,45,45,45)
print(check(tuple1))
```
