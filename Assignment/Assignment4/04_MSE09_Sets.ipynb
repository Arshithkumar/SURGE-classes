# Sets Assignment

```python
sampleset={"Yellow", "Orange", "Black"}
samplelist=["Blue", "Green", "Red"]
sampleset.update(samplelist)
print(sampleset)
```

```python
set1={10, 20, 30, 40, 50}
set2={30, 40, 50, 60, 70}
print(set1.intersection(set2)) #intersection()[gives the common values from both the sets]
```

```python
set1={10, 20, 30, 40, 50}
set2={30, 40, 50, 60, 70}
print(set1.union(set2))  #union()
```

```python
set1={10, 20, 30}
set2={20, 40, 50}
set1.difference_update(set2)
print("set1 ",set1)
```

```python
set1={10, 20, 30, 40, 50}
set1.difference_update({10,20,30})
print(set1)
```

```python
set1={10, 20, 30, 40, 50}
set2={30, 40, 50, 60, 70}
print(set1.symmetric_difference(set2)) #symmetric difference()[gives the unique values present in both the sets]
```

```python
set1={10, 20, 30, 40, 50}
set2={60, 70, 80, 90, 10}
x=set1.intersection(set2)
print("two sets have items in common","\n",x)    #we can also use isdisjoint() method to check if sets have common elements
```

```python
set1={10, 20, 30, 40, 50}
set2={30, 40, 50, 60, 70}
set1.symmetric_difference_update(set2)
print(set1)
```
