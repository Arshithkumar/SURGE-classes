```python
# Q1) Given two integer numbers return their product. If the product is greater than 1000, then return their sum
num1=int(input("enter the first number:"))
num2=int(input("enter the second number:"))
def multiply_or_sum(num1, num2):
    product = num1 * num2
    if product > 1000:
        return num1 + num2
    else:
        return product
result = multiply_or_sum(num1,num2)
print(result)
```

```python
# Q2) Given a range of the first 10 numbers, iterate from the start number to the end number, and in
# each iteration print the sum of the current number and previous number

# Define the range of numbers
start_num = 1
end_num = 10

# Initialize a variable to store the previous number
prev_num = 0

# Print the header message
print("Printing current and previous number sum in a range ({})".format(end_num))

# Iterate over the range of numbers and print the sum of the current and previous number
for num in range(start_num, end_num+1):
    sum_num = num + prev_num
    print("Current Number {} Previous Number {} Sum: {}".format(num, prev_num, sum_num))
    prev_num = num
```

```python
# Q4) Given a string and an integer number n, remove characters from a string starting from zero up to n and return a new string
# For example, removeChars("surgeclasses", 4) so output must be eclasses.

def removeChars(input_str, n):
    new_str = input_str[n:]
    return new_str
input_str = str(input("enter the string:"))
n = 4
new_str = removeChars(input_str, n)

# Print the result
print("Original String:", input_str)
print("New String after removing first {} characters:".format(n), new_str)
```

```python
# Q5) Given a list of numbers, return True if first and last number of a list is same
def checkFirstLast(lst):
    first = lst[0]
    last = lst[-1]
    if first == last:
        return True
    else:
        return False

lst1 =list(input())
print("List 1:", lst1, "Result:", checkFirstLast(lst1))
```

```python
# 10) Given a two list of numbers create a new list such that new list should contain only odd numbers from the first list and even numbers from the second list
def odd_from_list1_even_from_list2(list1, list2):
    result = []
    for num in list1:
        if num % 2 != 0: # odd number
            result.append(num)
    for num in list2:
        if num % 2 == 0: # even number
            result.append(num)
    return result

list1 = [10, 20, 23, 11, 17]
list2 = [13, 43, 24, 36, 12]
result_list = odd_from_list1_even_from_list2(list1, list2)
print(f"Result List is {result_list}")
```

```python
# Q12) Calculate income tax for the given income by adhering to the below rules
# Taxable Income Rate (in %)
# First $10,000 0
# Next $10,000 10
# The remaining 20

def calculate_income_tax(income):
    if income <= 10000:
        tax = 0
    elif income <= 20000:
        tax = (income - 10000) * 0.1
    else:
        tax = 10000 * 0.1 + (income - 20000) * 0.2
    return tax
taxable_income = 45000
income_tax = calculate_income_tax(taxable_income)
print("Income tax payable: $" + str(income_tax))
```
