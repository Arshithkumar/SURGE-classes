```python
#1) Accept two numbers from the user and calculate multiplication
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
result = num1 * num2
print(result)
```

```python
#2) Display “My Name Is James” as “My**Name**Is**James” using output formatting of
#a print() function
text = "My Name Is James"
words = text.split()
result = "**".join(words)
print(result)
```

```python
#3) Convert decimal number to octal using print() output formatting
def DecimalToOctal(n):
    if(n > 0):  
        DecimalToOctal(int(n/8))
        print(n%8, end='')
num= int(input("Enter the number: "))
DecimalToOctal(num)
```

```python
#4) Display float number with 2 decimal places using print()
# Define a float number
my_float = float(input("Enter the nmber:"))

# Use print() output formatting to display the float number with 2 decimal places
print("The float number is: {:.2f}".format(my_float))
```

```python
#5) Accept a list of 5 float numbers as an input from the user
my_list = []

for i in range(5):
    num = float(input("Enter a float number: "))
    # Append the float number to the list
    my_list.append(num)

# Print the list
print("The list of float numbers is:", my_list)
```

```python
#6) Write all content of a given file into a new file by skipping line number 5
with open('test.txt', 'r') as old_file:
    with open('new_file.txt', 'w') as new_file:
        for i, line in enumerate(old_file):
            if i != 4:  
                new_file.write(line)
```

```python
#7) Accept any three string from one input() call
input_str = input("Enter three strings:")
input_list = input_str.split()
string1, string2, string3 = input_list
print( string1, string2, string3)
```

```python
#8) Format the following data using a string.format() method.
#Given:
#totalMoney = 1000
#quantity = 3
#price = 450
totalMoney = 1000
quantity = 3
price = 450
output_str = "I have {money} dollars so I can buy {num_items} football for {item_price:.2f} dollars."
formatted_str = output_str.format(money=totalMoney, num_items=quantity, item_price=price)
print(formatted_str)
```

```python
#9) How to check file is empty or not
import os

filename = 'output.txt'

if os.path.getsize(filename) == 0:
    print(f"{filename} is empty.")
else:
    print(f"{filename} is not empty.")
```
