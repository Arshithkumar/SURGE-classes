```python
#1
def print_name_age(name, age):
    print("Name:", name)
    print("Age:", age)
print_name_age("Nandhu", 20)
```

```python
#2
def func1(*args):
    for arg in args:
        print(arg)
func1(20, 40, 60)
```

```python
func1(80, 100)
```

```python
#3
def calculation(a, b):
    addition = a + b
    subtraction = a - b
    return addition, subtraction
res = calculation(40, 10)
print(res)
```

```python
#4
def showEmployee(name, salary=9000):
    print("Employee Name:", name)
    print("Salary:", salary)
showEmployee("Krishna")
showEmployee("Swamy", 12000)
```

```python
#5
def outer_function(a, b):
    def inner_function(a, b):
        return a + b
    addition = inner_function(a, b)
    return addition + 5
result = outer_function(10, 20)
print(result)
```

```python
#6
def recursive_sum(n):
    if n == 0:
        return 0
    else:
        return n + recursive_sum(n-1)
result = recursive_sum(10)
print(result)
```

```python
#7
def displayStudent(name, age):
    print(name, age)
showStudent = displayStudent
showStudent("Chimtu", 16)
```

```python
#8
even_numbers = [num for num in range(4, 31) if num % 2 == 0]
print(even_numbers)
```
