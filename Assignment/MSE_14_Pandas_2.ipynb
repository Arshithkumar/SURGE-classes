```python
import pandas as pd

# Q1.Create the DataFrame
test_df = pd.DataFrame({'location': ['Bangalore', 'Mumbai', 'Hyderabad', 'Bangalore', 'Mumbai', 'Hyderabad'], 
                        'Sales': [1000,2000,3000,4000,5000,6000]}, 
                       columns=['location', 'Sales'])

# I. Identify Locationwise sales
locationwise_sales = test_df.groupby('location')['Sales'].sum()
print(locationwise_sales)

# II. Identify Locationwise avg. sales
locationwise_avg_sales = test_df.groupby('location')['Sales'].mean()
print(locationwise_avg_sales)

# III. Identify test_df.mean() and what does this value mean?
df_mean = test_df.mean()
print(df_mean)

# IV. select data belonging to locations greater than 3000 avg. sales
high_avg_sales = test_df.groupby('location').filter(lambda x: x['Sales'].mean() > 3000)
print(high_avg_sales)

# V. select data where sales greater than 3000
high_sales = test_df[test_df['Sales'] > 3000]
print(high_sales)
```

```python
import pandas as pd

# Q2.Create the DataFrame
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'], 'data': range(6)}, columns=['key', 'data'])

# I. value counts of 'key' column
print(df['key'].value_counts())

# II. groupby on 'key' column
print(df.groupby('key'))

# III. groupby on 'key' column and sum 'data' column
print(df.groupby('key').sum())

# IV. groupby on 'key' column and count rows
print(df.groupby('key').count())

# V. mean of DataFrame along rows (axis=0)
print(df.mean(axis="rows"))

# VI. mean of DataFrame along columns (axis=1)
print(df.mean(axis=0))

# VII. mean of all numeric columns in DataFrame
print(df.mean())

# VIII. groupby on 'key' column and calculate mean of 'data' column for each group
print(df.groupby('key').mean())

# IX. groupby on 'key' column and select the first row of each group
print(df.groupby('key').first())

# X. groupby on 'key' column and calculate mean absolute deviation of 'data' column for each group
print(df.groupby('key').mad())
```

```python
import numpy as np
import pandas as pd

# Q3.Create the DataFrame
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'], 
                   'data1': range(6), 'data2': rng.randint(0, 10, 6)}, 
                  columns = ['key', 'data1', 'data2'])

# I. get min values of data1 and data2 group wise
print(df.groupby('key').min())

# II. get min, max, sum and median values of data1 and data2 group wise
print(df.groupby('key').agg(['min', 'max', 'sum', 'median']))

# III. get min of data1 and max of data2 group wise
print(df.groupby('key').agg({'data1': 'min', 'data2': 'max'}))
```

```python
#Q4.Explain the functionality of following statements
#I. df.groupby('key').filter(lambda x: x['data2'].mean() < 4)
#II. df.groupby('key').filter(lambda x: x['data2'].mean() <= 4)

The filter() method in pandas is used to subset the groups based on some condition.

In both statements, groupby() method is applied to the DataFrame df to group the data by the column 'key'. 
Then, filter() method is used to subset the groups based on a condition specified as a lambda function.

In the first statement, the lambda function lambda x: x['data2'].mean() < 4 calculates the mean of 'data2' column for each group, 
and filters only those groups where the mean value is less than 4. So, the resulting DataFrame will contain only the rows where
the group's mean value of 'data2' is less than 4.

In the second statement, the lambda function lambda x: x['data2'].mean() <= 4 also calculates the mean of 'data2' column for
each group, but includes groups where the mean value is less than or equal to 4. So, the resulting DataFrame will contain only
the rows where the group's mean value of 'data2' is less than or equal to 4.
```

```python
#Q5.Execute df.groupby('key').filter(lambda x: x['data1'].std() > 4) and explain the output

The code df.groupby('key').filter(lambda x: x['data1'].std() > 4) performs the following steps:

1.Group the rows of the DataFrame df by the values in the column 'key'.
2.For each group, calculate the standard deviation of the values in the 'data1' column using the std() method.
3.Keep only the groups where the standard deviation of 'data1' is greater than 4, using the filter() method.
The resulting DataFrame will contain only those rows where the corresponding group has a standard deviation greater than 4 
for the 'data1' column. This means that the output DataFrame will only contain rows where the values in the 'data1' column are 
more spread out across the group, indicating a higher degree of variability or diversity in the data.

Therefore, the output of df.groupby('key').filter(lambda x: x['data1'].std() > 4) will be a DataFrame with the same columns as 
df, but only with the rows where the standard deviation of 'data1' for the corresponding group is greater than 4.
```

```python
#Q6.Execute df.groupby('key').transform(lambda x : x + x.mean())

The code df.groupby('key').transform(lambda x : x + x.mean()) performs the following steps:

1.Group the rows of the DataFrame df by the values in the column 'key'.
2.For each group, calculate the mean value of each column using the mean() method.
3.Subtract the mean value from each element in the group, using the transform() method with a lambda function.
The resulting DataFrame will have the same shape as df, but with the values in each group replaced by the original value
plus the mean value of the corresponding group. This means that each element in the resulting DataFrame will be shifted towards
the mean of its corresponding group.
```
