```python
# Q1) Create a series ‘S’ with values [5, 6, 7, 8, 9] and Print 
#I. Values of S?
#II. Indexes of S?
#III. S[1]
#IV. Type (S) 
import pandas as pd
S = pd.Series([5, 6, 7, 8, 9])
print(S)
print(S.index)
print(S[1])
print(type(S))
```

```python
# Q2) Create a series ‘T’ with values [7,8,9,10,11,12] and index of ['abi','def','ghi','jik','kjh','qwr'] Print 
# I. Values of S?
# II. Indexes of S?
# III. T[1]
# IV. T[‘abi’] and T[‘kjh’]
# V. T['def': 'kjh']
# VI. Update T[‘kjh’] to 200
# VII. Add T[‘mnc’] as 2000
import pandas as pd
T = pd.Series([7, 8, 9, 10, 11, 12], index=['abi', 'def', 'ghi', 'jik', 'kjh', 'qwr'])
print(T)
print(T.index)
print(T[1])
print(T['abi'])
print(T['kjh'])
print(T['def':'kjh'])
T['kjh'] = 200
print(T)
T['mnc'] = 2000
print(T)
```

```python
# Q6) create s1 = pd.Series([1,2,3,4]) and s2 = pd.Series(['a','b','c','d']) and then create a dataframe 
# with ‘name’ column and s2 as values and ‘age’ column and s1 as values
import pandas as pd

s1 = pd.Series([1, 2, 3, 4])
s2 = pd.Series(['a', 'b', 'c', 'd'])

df = pd.DataFrame({'name': s2, 'age': s1})
print(df)

```

```python
# Q7) create s1 = pd.Series([1,2,3,4]) and s2 = pd.Series(['a','b','c','d'],index=[4,5,6,7]) and then 
# create a dataframe with ‘name’ column and s2 as values and ‘age’ column and s1 as values
import pandas as pd

s1 = pd.Series([1, 2, 3, 4])
s2 = pd.Series(['a', 'b', 'c', 'd'], index=[4, 5, 6, 7])

df = pd.DataFrame({'name': s2, 'age': s1})
print(df)
```

```python
# Q8) Create data = array([[5, 5, 2, 4, 5], [0, 0, 4, 5, 6], [3, 8, 3, 5, 4], [0, 9, 0, 9, 0],[4, 3, 7, 6, 5]]) and 
# create ‘df’ from data and column names as ['c1','c2','c3','c4','c5'] and ['r1','r2','r3','r4','r5']
# I. df[:2]
# II. df['c1']
# III. df['c1']['r1']
# IV. df[:: 2]
import pandas as pd
import numpy as np

data = np.array([[5, 5, 2, 4, 5], [0, 0, 4, 5, 6], [3, 8, 3, 5, 4], [0, 9, 0, 9, 0],[4, 3, 7, 6, 5]])

df = pd.DataFrame(data, columns=['c1', 'c2', 'c3', 'c4', 'c5'], index=['r1', 'r2', 'r3', 'r4', 'r5'])

print(df[:2])
print(df['c1'])
print(df['c1']['r1'])
print(df[::2])
```

```python
# Q10) 
import pandas as pd

area = pd.Series({'California': 423967, 'Texas': 695662, 'New York': 141297, 'Florida': 170312, 'Illinois': 149995})
popu = pd.Series({'California': 38332521, 'Texas': 26448193,'New York': 19651127, 'Florida': 19552860, 'Illinois': 12882135})

data = pd.DataFrame({'area':area, 'popu':popu})

# I. Select ‘area’ column
print(data['area'])

# II. Select ‘popu’ column
print(data['popu'])

# III. Create density column by doing popu / area operation
data['density'] = data['popu'] / data['area']
print(data['density'])

# IV. Create new column ‘zipcode’ with values [1,2,3,4,5]
data['zipcode'] = [1,2,3,4,5]
print(data)

# V. data.T
print(data.T)

# VI. data.iloc[2:4,:2]
print(data.iloc[2:4, :2])

# VII. data.iloc[3,2]
print(data.iloc[3, 2])

# VIII. data.loc[‘Florida’,’density’]
print(data.loc['Florida', 'density'])

# IX. data[‘density’][‘Florida’]
print(data['density']['Florida'])

# X. write a for loop to print every element of data
for i in data.index:
    for j in data.columns:
        print(f"{i} {j}: {data.loc[i,j]}")

# XI. data.columns
print(data.columns)

# XII. data.index
print(data.index)

# XIII. data.iloc[0,0]
print(data.iloc[0,0])

# XIV. data[‘area’][2]
print(data['area'][2])

# XV. data[‘popu’][‘Florida’]
print(data['popu']['Florida'])

# XVI. data.loc['Florida':'New York']
print(data.loc['Florida':'New York'])

# XVII. data.loc['New York':'Florida']
print(data.loc['New York':'Florida'])

# XVIII. select dataframe when density > 100
print(data[data['density'] > 100])

# XIX. print zipcode where density > 100
print(data[data['density'] > 100]['zipcode'])

# XX. data[1:3]
print(data[1:3])
```
