```python
#function adding three numbers

def add(num1,num2,num3):
    return num1 + num2 + num3
```

```python
#function multiplying three numbers

def mul(num1,num2,num3):
    return num1 *num2 * num3
```

```python
l1 = [10,20,30,40,50]
l2 = [1,2,3,4,5,6,7,8]
l3 = [100,200,300,400,500,600]
m = map(add,l1,l2,l3)
print(list(m))
```
