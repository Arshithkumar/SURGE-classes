```python
def vowel_collector(given_string):
    vowel_string=[]
    for i in given_string:
        if i in 'aeiou':
            vowel_string.append(i)
    out_put=''.join(vowel_string)
    return out_put

out_string=vowel_collector('Sachin Ramesh Tendulkar (/ˌsʌtʃɪn tɛnˈduːlkər/ ( listen); born 24 April 1973) is an Indian former international cricketer who served as captain of the Indian national team. He is widely regarded as one of the greatest batsmen in the history of cricket.')
```
